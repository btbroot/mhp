# Inspector's hard job

You're an OSCE observer unit who's assignment is to confirm the presence of a certainly spotted Russian heavy artillery battery in Donetsk Oblast, Ukraine. There were some equally doubtfull reports suggesting that it might be located near the cities of either Avdiivka (A), Bilytske (B) or Chasiv Yar (C), so the exact location is yet unknown. But you've got not enough resources to conduct your inspection in all these distant places. In fact, you can only search one of them for evidences.

In these circumstances, you're afraid, you can't do any better than just guess randomly, choose one city out of three and hope that you'll be worth you salt. And naturally enough, you're going to find it with the probability of 1/3 (33%).

OK, you go with that. You're making your choice, and you're announcing it officially in the news insuring against any smart-ass moves, like: "Oh, well, we blew the whole place up, indeed, for agricultural needs, who knew that these guys were sniffing something out over there at that moment?! Alas, nobody's survived".

> Get some friends. Friends help ― Majority the Trash Heap, Fraggle Rock

Next morning, while wrapping up your stuff, you're being called by one of your precious local informers living in one of two remaining cities, which you don't plan to check this time. The informer's stating, there's nothing in his city.

Well, OK then, you say. Very kind of you, you say. I'm heading to a totally different place, as you might know from the news. And if you don't follow them, you'd better start to, and stop taking my time for nothing, I'm kind of in a hurry right now, thanks, bye.

Right?

No.

# Use your intuition, they say

> “Don't try to comprehend with your mind. Your minds are very limited. Use your intuition.”
― Madeleine L'Engle, A Wind in the Door 

Oh, thank you, Madeleine, for such a good advise! Let's indeed act intuitively and just stay whith the initial choice and follow your plan, since your friend didn't say anything important, you already have opted not to visit his city, anyways, it doesn't change a bit.

```python
def stay(initial_choice=None, **kwargs):
    # - You may change your mind, honey?
    # - No.
    return initial_choice
```

Now, let's solve the problem.

```python
from random import sample as rsample

def mhp(num_choices, strategy):
    choices = set(range(num_choices))
    prize_is_here = set(rsample(choices, 1))
    will_check = set(rsample(choices, 1))
    wont_check = choices.difference(will_check)
    # Informers never reveal the prize
    no_prize = wont_check.difference(prize_is_here)
    # Informers never reveal all
    revealed = set(rsample(no_prize, len(wont_check) - 1))
    unrevealed = choices.difference(revealed)
    # Time to apply your educated strategy and draw the final decision where to go
    final_choice = strategy(initial_choice=will_check, unrevealed=unrevealed)
    return final_choice == prize_is_here
```

And we need a litle wrapper in order to solve the problem a number of times and to plot the results: the rate of success.

```python
%pylab inline

def plotmhp(strategy=None, num_times=100000, num_choices=3):
    plot(cumsum([mhp(num_choices, strategy) for i in xrange(num_times)]))
    xlabel('rounds')
    ylabel('successfulness')
    ylim(xlim())
    grid(True)
```

```python
plotmhp(stay)
```

Here you go, your steady 33% of success, and you deserve it, stubborn. Thanks, Madeleine!

# Go to a blind date, they say

> "What are this blind dates?" he asked curiously.
"An old American institution of mismating."
― Rona Jaffe, The Best of Everything 

On second tought you realizes that after your friend has given you his priceless hint, the sutiation is different now. Originally, there were 3 potential battery locations to check out, and now it's only 2 to go. If you'd make your random guess now, choose one city out of two to inspect, then, naturally, you'd be going to find the battery with the probability of 1/2 (50%). 

That's much better! Thanks, buddy! Yeah! What a hell! You just have no idea which one, but it must be one of them! Let's update your strategy with tossing a fair coin and go blindly, weeee!

```python
def coin(unrevealed=None, **kwargs):
    # Immediately forget the initial choice and toss the coin
    return set(rsample(unrevealed, 1))
```

```python
plotmhp(coin)
```

Here you go, your 50%, if you insist, mismating hysteric.

# Waaait a minute!

> Who wears the biggest hat in the army? The one with the biggest head. ― Supposedly, a joke

If you, please, just hold your horses for a second and think...

When you're making your initial choice, you're splitting the original choices into exaclty two groups:

1. The group of your choice. There's only one city. You plan to go there.
2. The remaining group. All other cities are belonging to it. Let's remove them from the Earth's surface and place them on Mars, for the sake of clarity. The residents will be fine over there.

OK, so the probability of winning the prize with your group of choice is 1/3. And the corresponding probability of losing the prize with your group of choice is 2/3.

Do you understand that this 2/3 of a chance is on Mars, actually? No matter, how you'd make your initial choice, you always get a chance of 1/3. And the remaining chance of 2/3 always gets to Mars.

You would choose Mars, if you aren't dumb, since it's a bigger chance, but it isn't feasible: the group is too large (two cities), and you're limited to one city.

Conveniently, your friend have just made that big group on Mars feasible to choose, by revealing one city which has nothing. There's only one city left on Mars, and you can choose this group now, it's begging you!

```python
def reconsider(initial_choice=None, unrevealed=None):
    # Immediately change our mind
    return unrevealed.difference(initial_choice)
```

```python
plotmhp(reconsider)
```

Now you're starting to make right choices after all. But in case you aren't, here is the same problem with a hundred of choices instead of three.

```python
plotmhp(reconsider, num_choices=100)
```
